```python
import cv2 
import pytesseract 
from gtts import gTTS
from googletrans import Translator
import playsound
import os
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files (x86)\Tesseract-OCR\tesseract.exe"
img = cv2.imread("F:\work\OCR\img\sample.jpg")  
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)  
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 5)) 
dilation = cv2.dilate(thresh1, rect_kernel, iterations = 1) 
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE) 

im2 = img.copy()  
text1 = ""
sorted_ctrs = sorted(contours, key=lambda ctr: cv2.boundingRect(ctr)[1])

for cnt in sorted_ctrs:
    x, y, w, h = cv2.boundingRect(cnt)
    rect = cv2.rectangle(im2, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cropped = im2[y:y + h, x:x + w]
    text = pytesseract.image_to_string(cropped)
    text1 += text
# print(text1)
# cv2.imwrite(r"F:\work\test\images\rect1.jpg",rect)
tts = gTTS(text = text1, lang='en',slow=False)
tts.save(r"F:\work\OCR\img\text1.mp3")
playsound.playsound(r"F:\work\OCR\img\text1.mp3")
# toSpeak = gTTS(text = output, lang ='en', slow = False) 
# 	file = +"text1.mp3" 
# 	toSpeak.save(file)  
# 	playsound.playsound(file, True) 
    
#     text = “ This is in english language” 
# var = gTTS(text = text,lang = ‘en’) 
# var.save(‘eng.mp3’) playsound(‘.\eng.mp3’)

# engine = pyttsx3.init() 
# voices = engine.getProperty('voices')
# engine.setProperty('voice', voices[1].id) 
# engine.setProperty('rate', 150)
# an audio will be played which speaks the test if pyttsx3 recognizes it 
# engine.say(k)
# engine.runAndWait() 
```
